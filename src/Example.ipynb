```python
filas = joblib.Parallel(n_jobs=-1)(joblib.delayed(downloader.recopilar_noticias)(urls) for urls in tqdm(json_df['url']))
filas
```

```python
# Añadimos las filas al dataframe, son columnas que antes no existían:
for i in range(len(filas)):
    json_df.at[i,'autores'] = ";".join(filas[i]['autores'])
    json_df.at[i,'keywords'] = ";".join(filas[i]['keywords'])
    json_df.at[i,'summary'] = filas[i]['summary']
    json_df.at[i,'fecha'] = filas[i]['fecha']
    json_df.at[i,'cuerpo'] = filas[i]['cuerpo']
```

```python
def pipeline_total(query, positives, negatives, start_date, end_date):
    final_query = downloader.query_completion(downloader.name_amplifier(downloader.query_cleaner(query)), positives, negativos)
    url = downloader.api_url_constructor(f'{final_query}')
    json_noticias = downloader.get_json(url, query, download = False)
    json_df = downloader.json_df(json_noticias)
    filas = joblib.Parallel(n_jobs=-1)(joblib.delayed(downloader.recopilar_noticias)(urls) for urls in tqdm(json_df['url']))
    for i in range(len(filas)):
        json_df.at[i,'autores'] = ";".join(filas[i]['autores'])
        json_df.at[i,'keywords'] = ";".join(filas[i]['keywords'])
        json_df.at[i,'summary'] = filas[i]['summary']
        json_df.at[i,'fecha'] = filas[i]['fecha']
        json_df.at[i,'cuerpo'] = filas[i]['cuerpo']
    return json_df
```

```python
pipeline_total(query, positives, negativos, start_date, end_date)
```
